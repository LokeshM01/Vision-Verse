```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage.io import imshow, imread
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
```

```python
k=5
```

```python
coffee = data.coffee()
```

```python
plt.figure(1)

plt.axis=("off")
plt.title("Original image")
plt.imshow(coffee)
```

```python
pixels=coffee.reshape(-1,3)
```

```python
histogram, bins = np.histogram(pixels, bins=256, range=(0, 256))
```

```python
image_array_sample=shuffle(pixels,random_state=0)[:1000]
kmeans=KMeans(n_clusters=k).fit(image_array_sample)
labels=kmeans.predict(pixels)
```

```python
image_out=np.zeros(coffee.shape)
label_idx=0
for i in range(coffee.shape[0]):
  for j in range(coffee.shape[1]):
    image_out[i][j]=kmeans.cluster_centers_[labels[label_idx]]
    label_idx += 1

```

```python
image_out_uint8 = image_out.astype(np.uint8)
```

```python
plt.figure()

plt.title("Quantised image image")
plt.imshow(image_out_uint8)
plt.show()
```
