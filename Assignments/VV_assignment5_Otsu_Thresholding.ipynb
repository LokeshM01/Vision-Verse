Perform Otsu thresholding on the following image 'boat.jpg' and show the final output. Also, perform normal thresholding and compare the two results.

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
image=cv2.imread("boat.jpg")
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(image_gray)
```

```python
th=0;
```

```python
threshold,x=cv2.threshold(image_gray,th,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
print(threshold)
```

```python
cv2_imshow(x)
```
