```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("original.jpg")
```

```python
cv2_imshow(img)
```

# ***Gaussiann filter***

```python
blur_img = cv2.GaussianBlur(img, (9,9), 0)

cv2_imshow(blur_img)
```

# ***Median filter***

```python
median = cv2.medianBlur(img,9)
cv2_imshow(median)
```

# ***Bilateral filter***

```python
blur = cv2.bilateralFilter(img,9,85,85)
cv2_imshow(blur)
```

# ***Laplacian filter***

```python

laplacian = cv2.Laplacian(img, cv2.CV_64F)

cv2_imshow(laplacian)
```

# ***Box filter / Averaging***

```python
blur = cv2.blur(img,(7,7))
cv2_imshow(blur)
```
