```python
 Perform edge detection using following techniques on all images given and conclude which
 method gives best results:
 1) Canny Edge Detection
 2) Hough Transform
 3) Laplacian
 4) Edge Detection Using Gradients-Sobel,Prewitt,Robert Operators
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("3.jpeg", cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(img, cv2.CV_64F, ksize=3)
lap = np.uint8(np.absolute(lap))
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
edges = cv2.Canny(img,100,200)
```

```python
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
```

```python
sobelCombined = cv2.bitwise_or(sobelX, sobelY)
```

```python
titles = ['image', 'Laplacian', 'sobelX', 'sobelY', 'sobelCombined', 'Canny']
images = [img, lap, sobelX, sobelY, sobelCombined, edges]
```

```python
for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
